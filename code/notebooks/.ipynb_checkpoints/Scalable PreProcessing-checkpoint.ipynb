 ## Preprocessing Steps

### Identify all images in the library

Since I last worked on this, I got a NAS and have all my images there, so now I'll have to be reading and writing files through my local network instead of a connected external hard drive.

### Clean up any file names so we have a consistent convention

For some reason, I changed the formatting of timestamps in the filenames throughout the data collection time. Let's iterate over the files and rename to a standard timestamp filename: the same that we've constructed in image_times_mst.

### Collect all the timestamps from the filenames

I'll need these timestamps for a number of purposes, so it makes sense to pull them out and maintain the same ordering as all the images in the library

### Crop the images

Cropping to a set extent ensures we get the same area for all images, that they could be overlaid in the future, and that the changes between images are only color-based, not extent-based.

### Count the colors and how many pixels are assigned each color

The full dataset should have a timestamp and the colors expressed at that point. We can melt that data and have it in a tall format or keep it as a relationally-structured dataset. I'll probably do both but it will be the most complete form of this data outside the original images.

## Class with various methods to process newly added images

```python
class screenshot_handler:
    def __init__(self):
        self.input_dir = None
        self.image_parent_directory = None
        self.interim_dir_exists = False
        self.output_dir_exists = False
        self.image_list = []
        self.renamed_images = []
        self.cropped_images = []
        self.image_times = None
        self.image_times_str = None
        self.rows_dict = None
        from collections import Counter
        import pandas as pd
        from PIL import Image
        from tqdm import tqdm
        from pathlib import Path
        import re
        import shutil

    def create_structure(self, input_dir = None):
        # check for valid path
        if input_dir is None:
            print('valid input directory must be provided')
        elif os.path.isdir(input_dir) == False:
            print(f'input directory,\n{input_dir}\n not recognized as valid path')
        else:
            #replace escaping backslash path convention
            input_dir = input_dir.replace('\\', '/')
            
            # set class variable
            self.input_dir = input_dir
            
            # create file structure for images
            os.makedirs(input_dir + 'imageprocessing', exist_ok=True)
            os.makedirs(input_dir + 'imageprocessing/renamed', exist_ok=True)
            os.makedirs(input_dir + 'imageprocessing/cropped', exist_ok=True)
    
    
    def collect_image_paths(self, filetype='.png', image_parent_directory = None):
        
        if image_parent_directory is None:
            print('no valid input directory has been specified. Attempting to use parent directory specified in screenshot_handler.create_structure')
            
            # no impage parent directory provided, set to input directory
            image_parent_directory = self.input_directory
            
        if os.path.isdir(image_parent_directory) == False:
            print(f'parent directory for images,\n{image_parent_directory}\n not recognized as valid directory')
            
        if image_parent_directory[-1] != "/" and image_parent_directory[-1] != "\\": 
            image_parent_directory = image_parent_directory + "/"
            self.image_parent_directory = image_parent_directory
            
        
        # get all images and their make times
        search_logic = image_parent_directory + "raw_files/**/*" + filetype
        print(f'searching using glob pattern {search_logic}')
        
        # collect image list by order of file creation and sort ascending by creation time
        image_list = sorted(glob.glob(search_logic, recursive=True), key=os.path.getmtime)
        
        # assign list to class to store for future use
        self.image_list = image_list
        
        print(f'identified {len(self.image_list)} images and {len(set(self.image_list))} unique images')

        
        
    def strip_out_file_timestamps(self, strings_to_remove_from_filenames = ['denvertraffic', '.png'], time_str = "%Y%m%d%H%M%S"):

        # copy list to avoid issues if a file throws an error in the process
        loop_image_times = self.image_list

        # get the timestamps from each file by cutting away pieces of the filepath strings
        print(f'Removing substrings from filenames {strings_to_remove_from_filenames}')
        for substring in strings_to_remove_from_filenames:
            loop_image_times = [Path(x).stem.replace(substring,'') for x in tqdm(loop_image_times)]
            
            
        # keep only the integers in the filename and nothing else
        print('Removing all non-integer characters from filenames')
        loop_image_times = [re.sub('[^0-9]', '', x) for x in tqdm(loop_image_times)]

         #change to datetime objects
        print(f'Converting numeric filename remnants to datetimes using pattern {time_str}')
        loop_image_times = [datetime.strptime(x,time_str) for x in tqdm(loop_image_times)]
        
        # add as class variable
        self.image_times = loop_image_times
        
        if len(self.image_times) != len(self.image_list):
            print(f"""Warning: The length of the image times is {len(self.image_times)} and does not match the length of the
            image list which has len {len(self.image_list)}""")
            
            
    def copy_processed_files(self, file_extension = '.png', output_timestamp_format = "%Y_%m_%d_%H_%M_%S_UTC"):

        #rename all photos that were found in via glob to just the timestamp filenames
        
        # create new list of filepaths for write that correspond with the self.image_list files
        filenames = [datetime.strftime(x, output_timestamp_format) + file_extension for x in self.image_times]
        new_filepaths = [self.input_dir + "/imageprocessing/renamed/" + x for x in filenames]
        
        # loop through image list, check if corresponding file exists
        # if new file doesnt exist, create it using the corresponding position
        print('Validating / Copying Files')
        for pos, file in tqdm(enumerate(self.image_list)):
            if os.path.isfile(new_filepaths[pos]):
                continue
            else:
#                 shutil.move(file, new_filepaths[pos])
                shutil.copy(file, new_filepaths[pos])

        
        # add sorted images for future steps
        print('Files renamed. Collecting list of files for next step')
        
        # add filenames alone to class variable
        self.image_times_str = [datetime.strftime(x,output_timestamp_format) for x in self.image_times]
    
    def crop_images(self, extent=(500, 0, 1720, 1080), file_extension = '.png'):
        from PIL import Image
        print('Collecting all renamed images from imageprocessing/renamed/ in preparation to crop')
        self.renamed_images = sorted(glob.glob(self.input_dir + "imageprocessing/renamed/*" + file_extension, recursive=True), key=os.path.getmtime)

        print(f'Cropping each of {len(self.renamed_images)} images')
        for pos, image in tqdm(enumerate(self.renamed_images)):
            
            # if cropped_file_exists, continue. Else, save cropped image
            output_name = self.input_dir + "imageprocessing/cropped/" + self.image_times_str[pos] + file_extension
            if os.path.isfile(output_name):
                        continue
            else:
                try:
                    with Image.open(image) as screenshot:
                        # crop using rectangular bounding box defined in crop_extent
                        screenshot = screenshot.crop(box=extent)
                        screenshot.save(output_name)
                            
                except OSError as e:
                    print(f'OSError occurred: {e} on image {image}. Moving to next image')
                    
        print('Collecting all cropped images in imageprocessing/cropped/')
        self.cropped_images = sorted(glob.glob(self.input_dir + 'imageprocessing/cropped/*.png', recursive=True), key=os.path.getmtime)
            
    def count_pixel_tuples(self):
        from PIL import Image
        from collections import Counter
        from pathlib import Path


        # handle if file does not exist
        if  os.path.isfile('full_dataset_as_dict_2.txt'):
            with open('full_dataset_as_dict_2.txt', 'r') as f:
                print('loading in currently stored dataset')
                full_dataset = json.load(f)
        else:
            with open('full_dataset_as_dict_2.txt', 'a') as f:
                f.write(json.dumps(dict()))
                full_dataset = json.load(f)
                
        # redundancy: duplicated dictionary to capture file contents, remove in future
        rows_dict = {}
        
        # existing keys so we dont have to call them repeatedly
        full_dataset_keys = list(full_dataset.keys())
        
        # loop through cropped images and extract colors for every possible pixel combination
        print('Looping through cropped images to extract pixel counts in each image')
        for pos, filename in tqdm(enumerate(self.cropped_images)):
            if Path(filename).stem in full_dataset_keys:
                continue

            else:
                with Image.open(filename) as screenshot:

                    # extract list of pixel tuples
                    pixel_tuple_list = list(screenshot.getdata())

                    # convert to strings
                    pixel_tuple_list = [str(x) for x in pixel_tuple_list]

                    # use counter to get counts per pixel
                    pixel_counts = Counter(pixel_tuple_list)

                    # add record to rows dictionary
                    timestamp = self.image_times_str[pos]
                    rows_dict[timestamp] = pixel_counts
                    full_dataset[timestamp] = pixel_counts

        # store resulting dictionary
        print('Storing pixel counts')
                
        self.rows_dict = rows_dict
        with open('full_dataset_as_dict_2.txt', 'w') as convert_file: 
             convert_file.write(json.dumps(full_dataset))
        
        return full_dataset
```

```python
color_tuples = [ast.literal_eval(x) for x in rows_to_be[0].keys()]
```

```python
color_tuples[0]
```

```python
def rgba_to_hex(rgba):
    """
    converts rgba tuple to hexadecimal string
    """
    a, b, c, d = rgba
    result =  '#{:02x}{:02x}{:02x}{:02x}'.format( a, b , c, d )

    return result


def tuple_mean(rgba):
    """
    calculates mean of rgba values in tuple
    """
    a,b,c,d = rgba
    result = (a+b+c+d)/4.0
    return result
```

```python
#fit k-means on hsv data
kmeans = KMeans(n_clusters=20, random_state=9)

kmeans.fit(color_tuples)

# rgba to hex (tried using seaborn swarmplot using hex values but that was not rendering)
# Here, used to define color per point.
colors = [rgba_to_hex(x) for x in color_tuples]

#stores colors per label after clustering
color_dict = {}

```

```python
from matplotlib import pyplot as plt
#adds rgba values to dict under label
for pair in tqdm(zip(kmeans.labels_, color_tuples,)):
    if pair[0] in color_dict:
        color_dict[pair[0]].append(pair[1])
    else:
        color_dict[pair[0]] = []
        color_dict[pair[0]].append(pair[1])    

plot_df = pd.DataFrame()

plot_df['labels'] = kmeans.fit_predict(color_tuples)
# plot_df['hsv'] = color_tuples
plot_df['rgb'] = color_tuples
plot_df['mean'] = [tuple_mean(x) for x in color_tuples]


plt.figure(figsize=(20,12))
plt.scatter(plot_df['labels'], plot_df['mean'], s=150, edgecolor='', c=colors)
plt.xticks([i for i in range(len(kmeans.cluster_centers_))])
plt.ylabel('Mean pixel value')
plt.xlabel('Cluster ID')
plt.title('K-means clustering: Labels and colors')
plt.show()
```

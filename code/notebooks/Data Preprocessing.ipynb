### Generalize function to extract pixel dictionary from specified extent

```python
def time_series_from_target_area(pixel_dict,
                                 header=True,
                                 file_list=[],
                                 bounding_area=(),
                                 time_format='',
                                 hour_diff=-6,
                                 resample_by=''):
    """
    Creates pandas dataframe with pixel counts by unique pixel tuple values.
    Crops image by bounding area. Extacts timestamps from filenames,
    Resamples time series by 20 min right closed value intervals with any grouped images aggregated using .mean().
    """

    import re
    from PIL import Image
    from datetime import datetime, timedelta
    function_dict = {}

    #----------------------------

    if type(pixel_dict) == str:  #if passed filepath
        #read in text file of pixel labels and store as dictionary
        with open(pixel_dict, 'r') as inf:

            #skip first line that has a title
            if header:
                next(inf)

                #store text file dictionary in pixel_labels
                pixel_labels = ast.literal_eval(inf.read())
            else:
                pixel_labels = ast.literal_eval(inf.read())
    else:  #if passed a dictionary
        pixel_labels = pixel_dict

    #extract timestamps
    #keep only filename, strip alphabetic characters, replace file extension
    #get filename
    image_times = [os.path.basename(x) for x in file_list]

    #replace a-z A-Z characters in filename
    image_times = [re.sub('[a-zA-Z]', '', x) for x in image_times]

    #replace underscores with colons and periods with null
    image_times = [
        x.replace('.png', '').replace('_', ':').replace('.', '')
        for x in image_times
    ]

    #convert to datetime given format and then update to local time using hour_diff
    image_times = [datetime.strptime(x, time_format) for x in image_times]

    diff_hours = timedelta(hours=hour_diff)

    image_times_localized = [x + diff_hours for x in image_times]
    print('converted image timestamps to local time')

    #----------------------------
    #crop images
    #----------------------------

    print('extracting dictionaries from file list')
    for position, im in enumerate(tqdm(file_list)):
        im_temp = Image.open(im)

        #if bounding area provided
        if len(bounding_area) != 0:
            box = bounding_area
        else:
            w, h = im_temp.size
            box = (0, 0, w, h)

        im_temp = im_temp.crop(box)

        # to hold pixel tuples in loop
        image_dict = {}

        #extract all pixel tuples from cropped image
        pixel_tuple_list = list(im_temp.getdata())

        #parse each and update dictionary
        for pixel_tuple in pixel_tuple_list:
            if pixel_tuple in image_dict:
                image_dict[pixel_tuple] += 1
            else:
                image_dict[pixel_tuple] = 1

    #add entire image_dict to function_dict
        function_dict[position] = image_dict

    #create dataframe
    #create dataframe from dictionary with each record as a row
    time_series = pd.DataFrame.from_dict(function_dict, orient='index')

    #add datetime stamps in mst
    time_series['date_time'] = image_times_localized

    #reset index to timestamps for time indexing
    time_series.index = pd.to_datetime(time_series.date_time)
    del time_series['date_time']
    time_series = time_series.astype('float32')
    time_series.fillna(0, inplace=True)
    if resample_by != '':  # if value was provided for resampling
        resampled = time_series.resample(resample_by,
                                         label='right',
                                         closed='right').mean()
    else:
        resampled = time_series

    return resampled
```

# Categorizing Ranges of Color


With thousands of unique RGBA colors, I imagine we can get that down to a much-reduced space and then manually label the color groups within. 

First, I can loop through the image dataset and compare a set of pixel values to a a set from the denver map with no traffic enabled. This will isolate colors only used in the traffic layers. Then, we can process the images using only the colors only present in traffic layers and create some pixel labels from there.

### Creating a set of pixel labels from the basemap

```python
no_traffic = Image.open("../../data/class/no_traffic/denver2020-05-12 20_06_44.png")

no_traffic_pixel_list = list(no_traffic.getdata())
no_traffic_pixel_set = set(no_traffic_pixel_list)
```

### Create set with pixel tuples only present in traffic images

```python
image_list = sorted(glob.glob("../../data/class/full_images/*.png"), key=os.path.getmtime)
traffic_only_set = set([])

for pos, val in enumerate(tqdm(image_list)):
    temp_im = Image.open(val)
    temp_pixel_values = list(temp_im.getdata())
    set_temp = set(temp_pixel_values)
    diff = set_temp - no_traffic_pixel_set
    
    traffic_only_set.update(diff)
```

## Clustering Traffic Pixel Values

Now, traffic_only_set contains the pixel rgba values that appear in one of the traffic images but not in the map with no traffic layer enabled. at this point, the set is still larger than I would want to manually parse and assign text labels for easier downstream interpretation.

With these pixel values represented in in rgba (but where a is always 255 in this case), I'll convert to hsv for better clustering ability. See
http://www.cse.msu.edu/~pramanik/research/papers/2002Papers/icip.hsv.pdf for more context.

```python
# sourced from https://www.w3resource.com/python-exercises/math/python-math-exercise-77.php
def rgb_to_hsv(val):
    r, g, b, a = val
    r, g, b, a = (r/255.0, g/255.0, b/255.0, a/255.0)
    mx = max(r, g, b)
    mn = min(r, g, b)
    df = mx-mn
    if mx == mn:
        h = 0
    elif mx == r:
        h = (60 * ((g-b)/df) + 360) % 360
    elif mx == g:
        h = (60 * ((b-r)/df) + 120) % 360
    elif mx == b:
        h = (60 * ((r-g)/df) + 240) % 360
    if mx == 0:
        s = 0
    else:
        s = (df/mx)*100
    v = mx*100
    return (h, s, v)
```

I then convert the rgb values to hsv using the function defined and store as hsv_set

```python
hsv_set = [rgb_to_hsv(x) for x in traffic_only_set]
```

Now, I write these out as csv's to save intermediate output. That way, I don't have to recalculate all the intermediate products should I want to adjust fitting later on in a different session.

```python
#write out hsv tuples only present in traffic images
with open('../../data/class/color_lists/hsv_traffic_only_set.csv', "w") as output:
    writer = csv.writer(output, lineterminator='\n')
    for val in hsv_set:
        writer.writerow([val])
        
#write out rgba tuples only present in traffic images
with open('../../data/class/color_lists/rgba_traffic_only.csv', "w") as output:
    writer = csv.writer(output, lineterminator='\n')
    for val in traffic_only_set:
        writer.writerow([val])
            
```

### K - Means

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D



kmeans = KMeans(n_clusters=15, random_state=0).fit(list(hsv_set))

clusts = kmeans.fit_predict(hsv_set)
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(kmeans.cluster_centers_[:, 2],
            kmeans.cluster_centers_[:, 0],
            kmeans.cluster_centers_[:, 1],
            s = 100,
            marker='o',
            c='red',
            label='centroids')

scatter = ax.scatter([list(x)[2] for x in hsv_set],
                     [list(x)[0] for x in hsv_set],
                     [list(x)[1] for x in hsv_set],
                     c=clusts,s=5, cmap='winter')


ax.set_title('K-Means Clustering')
ax.set_xlabel('Value')
ax.set_ylabel('Hue')
ax.set_zlabel('Saturation')
ax.legend()
plt.show()
```

After wrestling with the matplotlib 3-d approach, I turned back to 2d visualization of the labels. I wanted to color each point by the actual pixel color so I could check out how the colors were separating. The classic scatterplot enters the game!

```python
# convert rgb to hex
'#{:02x}{:02x}{:02x}'.format( 120, 0 , 255 )
def rgba_to_hex(rgba):
    """
    converts rgba tuple to hexadecimal string
    """
    a, b, c, d = rgba
    result =  '#{:02x}{:02x}{:02x}{:02x}'.format( a, b , c, d )

    return result

def tuple_mean(rgba):
    """
    calculates mean of rgba values in tuple
    """
    a,b,c,d = rgba
    result = (a+b+c+d)/4.0
    return result
```

```python
import matplotlib.pyplot as plt
import numpy as np

#fit k-means on hsv data
kmeans = KMeans(n_clusters=30, random_state=0).fit(list(hsv_set))

# rgba to hex (tried using seaborn swarmplot using hex values but that was not rendering)
# Here, used to define color per point.
colors = [rgba_to_hex(x) for x in list(traffic_only_set)]

#stores colors per label after clustering
color_dict = {}

#adds rgba values to dict under label
for pair in zip(kmeans.labels_, traffic_only_set,):
    if pair[0] in color_dict:
        color_dict[pair[0]].append(pair[1])
    else:
        color_dict[pair[0]] = []
        color_dict[pair[0]].append(pair[1])    

plot_df = pd.DataFrame()
plot_df['labels'] = kmeans.fit_predict(hsv_set)
plot_df['hsv'] = list(hsv_set)
plot_df['rgb'] = list(traffic_only_set)
plot_df['mean'] = [tuple_mean(x) for x in traffic_only_set]


plt.figure(figsize=(20,12))
plt.scatter(plot_df['labels'], plot_df['mean'], s=150, edgecolor='', c=colors)
plt.xticks([i for i in range(len(kmeans.cluster_centers_))])
plt.ylabel('Mean pixel value')
plt.xlabel('Cluster ID')
plt.title('K-means clustering: Labels and colors')
plt.show()
```

The goal here is to identify some clusters that represent the traffic layers I'm interested in and extract all their pixel value tuples. With an high number of clusters, the colors get separated out more finely. With fewer, more colors are included together.

With that, I can actually a fairly high number of clusters but only extract a few labels from them.  Some clusters have varying shades of gray, bright blues, etc and aren't in the green-red space the traffic of interest uses.

Therefore, I'm using the clusters with rich reds, greens, and oranges, and crimsons for my layers of interest. I may mask the images using these colors to check that they are mainly represented by lines and not large polygonal or patchy extents (which would indicate basemap pixels are being included in the clustering).

```python
a = glob.glob('../../data/class/cluster_outputs/*')

for file in a:
    os.remove(file)
    
n = 747
a = Image.open(image_list[n])
a.save('../../data/class/cluster_outputs/original.png')


#loop through labels
for label in sorted(list(color_dict.keys())):

    #read in a chosen image
    im = plt.imread(image_list[n])

    #convert 0-1 to 0-255
    im = np.round(im * 255, 0)

    #store empty array with same dimension
    empty_img = np.zeros_like(im)

    #unpack array
    reds, greens, blues, alphas = im[:, :, 0], im[:, :, 1], im[:, :,
                                                               2], im[:, :, 3]

    #loop each color in label
    for color in tqdm(color_dict[label]):
        r, g, b, a = color

        #mask image and write black
        mask = ((greens == g) & (reds == r) & (blues == b))
        empty_img[mask] = (0, 0, 0, 255)

    #rescale to 255 space
    empty_img = empty_img / 255

    #show
    plt.figure(figsize=(16, 10))
    plt.imshow(empty_img, aspect='auto')
    plt.xlabel('pixel position')
    plt.ylabel('pixel position')
    plt.title('Label number: ' + str(label))

    #save with label
    plt.savefig(
        "../../data/class/cluster_outputs/" +
        str(label) + '.png')
    plt.close()
```
